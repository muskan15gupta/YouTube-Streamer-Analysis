# TREND ANALYSIS

```python
categories_distribution=df['Categories'].value_counts()
print(categories_distribution)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x=categories_distribution.values, y=categories_distribution.index)
plt.title('Distribution of Categories Among Top YouTube Streamers')
plt.xlabel('Number of Streamers')
plt.ylabel('Categories')
plt.show()
```

```python
category_engagement = df.groupby('Categories').agg({
    'Subscribers': 'mean',
    'Likes': 'mean',
    'Comments': 'mean',
}).reset_index()

category_engagement = category_engagement.sort_values(by='Subscribers', ascending=False)

print(category_engagement)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Subscribers', y='Categories', data=category_engagement)
plt.title('Average Subscribers per Category')
plt.xlabel('Average Subscribers')
plt.ylabel('Categories')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Likes', y='Categories', data=category_engagement)
plt.title('Average Likes per Category')
plt.xlabel('Average Likes')
plt.ylabel('Categories')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Comments', y='Categories', data=category_engagement)
plt.title('Average Comments per Category')
plt.xlabel('Average Comments')
plt.ylabel('Categories')
plt.show()
```

```python
category_engagement['Total Engagement'] = category_engagement['Subscribers'] + category_engagement['Likes'] + category_engagement['Comments']

category_engagement = category_engagement.sort_values(by='Total Engagement', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Total Engagement', y='Categories', data=category_engagement, color='blue', orient='h')
plt.title('Total Engagement per Category')
plt.xlabel('Total Engagement')
plt.ylabel('Categories')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def normalize_metric(metric):
    if metric.max() - metric.min() == 0:
        return metric  
    else:
        return (metric - metric.min()) / (metric.max() - metric.min())  

category_engagement['Normalized_Subscribers'] = normalize_metric(category_engagement['Subscribers'])
category_engagement['Normalized_Likes'] = normalize_metric(category_engagement['Likes'])
category_engagement['Normalized_Comments'] = normalize_metric(category_engagement['Comments'])

print(category_engagement[['Normalized_Subscribers', 'Normalized_Likes', 'Normalized_Comments']])
```

```python
import pandas as pd

weights = {'Subscribers': 0.5, 'Likes': 0.3, 'Comments': 0.2}

category_engagement['Weighted_Engagement_Score'] = (category_engagement['Subscribers'] * weights['Subscribers']) + \
                                                    (category_engagement['Likes'] * weights['Likes']) + \
                                                    (category_engagement['Comments'] * weights['Comments'])

category_engagement = category_engagement.sort_values(by='Weighted_Engagement_Score', ascending=False)

print(category_engagement[['Categories', 'Weighted_Engagement_Score']])
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Weighted_Engagement_Score', y='Categories', data=category_engagement, color='blue')
plt.title('Weighted Engagement Score per Category')
plt.xlabel('Weighted Engagement Score')
plt.ylabel('Categories')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```

```python
df2 = category_engagement.drop(columns=['Subscribers', 'Likes', 'Comments', 'Total Engagement'])
df2.head()                          
```

```python
from sklearn.metrics.pairwise import cosine_similarity

features = df2[['Normalized_Subscribers', 'Normalized_Likes', 'Normalized_Comments', 'Weighted_Engagement_Score']]
features = features.fillna(0)
```

```python
similarity_matrix = cosine_similarity(features)
similarity_df = pd.DataFrame(similarity_matrix, index=df2['Categories'], columns=df2['Categories'])
```

```python
def get_similar_categories(category, similarity_df, top_n=5):
    similarity_scores = similarity_df[category].sort_values(ascending=False)
    similar_categories = similarity_scores.drop(category).head(top_n)
    return similar_categories

similar_categories = get_similar_categories('Videojuegos, Humor', similarity_df)
print("Categories similar to 'Videojuegos, Humor':\n", similar_categories)
```
